```python
# list of dataframes containing all data given one field
dfs=[pd.read_excel(open('./raw/'+dirName[0]+'/'+filename, 'rb'),
                   usecols=[1,2],
                   names=['school', 'data'],
                   dtype={'school':np.object, 'data': np.float64})
 for filename in fileName]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['school'],
                                            how='outer'), dfs)
```
